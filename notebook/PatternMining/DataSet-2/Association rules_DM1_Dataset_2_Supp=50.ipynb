# Description of the Value

We have ecide to do some proves with different characterized value of support. This value are 2,10,20,30,40,50,60,70,80,90.
After that we have to do some proves with different type of target , a for frequent itemset, m for maximal itemset e c for closed itemset. 

For choosing the best combination we try to create the frequent itemset with 2/3/4 min value.

# Support = 50%

 Frequent Itemsets with min value 2/3/4

```python
itemsets = apriori(baskets, supp=50, report ='s', target='a',zmin = 2)
print('Number of itemsets:', len(itemsets))
itemsets
```

```python
itemsets1 = apriori(baskets, supp=50, report ='s', target='a',zmin = 3)
print('Number of itemsets:', len(itemsets1))
itemsets1
```
