```python
global_info = {}

# dict having the following format:
"""{"model_name": { "test_type": { "fpr": [],
                                "tpr": [],
                                "roc": int;  # area under ROC-curve
        }
    }
}"""
```

```python
# functions to save and load json file
import pickle
"""with open('knn_best_onehot_models.json', 'wb') as handle:
    pickle.dump(global_info, handle, protocol=pickle.HIGHEST_PROTOCOL)"""
    
with open('knn_best_numerical_models.json', 'rb') as handle:
    global_info = pickle.load(handle)
```

```python
def draw_multiple_roc_and_ap_curves(type_flag, colors):  # type_flag=validation or type_flag=test
    # draw_roc_auc
    i = 0
    plt.figure(figsize=(8, 5))
    for model_name in global_info.keys():
        color = colors[i]
        i += 1
        fpr = global_info[str(model_name)][type_flag]["fpr"]
        tpr = global_info[str(model_name)][type_flag]["tpr"]
        roc_auc = global_info[str(model_name)][type_flag]["roc"] 
        plt.plot(fpr, tpr, color=color, label='%s=%0.4f' % (model_name, roc_auc))
    plt.plot([0, 1], [0, 1], 'k--', color="red", label='no-skilled model=0.5000') 
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate') 
    plt.tick_params(axis='both', which='major')
    plt.legend(loc="lower right", title="AUC", frameon=True)
    plt.title("Models' %s ROC-curve" % type_flag)
    plt.show()
```

```python
colors = ["gold", "purple", "grey", "cyan", "lime", "red", "brown", "dodgerblue"]  
# plotting 8 models (for clarity maybe is better to do 4 and 4)
draw_multiple_roc_and_ap_curves("validation", colors)
```

```python
colors = ["gold", "purple", "grey", "cyan", "lime", "red", "brown", "dodgerblue"]  
# plotting 8 models (for clarity maybe is better to do 4 and 4)
draw_multiple_roc_and_ap_curves("test", colors)
```
